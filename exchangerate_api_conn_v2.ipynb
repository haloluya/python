```python
import requests
import json                 # Used to load data into JSON format
import pandas as pd
import cx_Oracle
from pprint import pprint   # pretty-print

request = requests.get("https://v6.exchangerate-api.com/v6/31ce24daeeb8712e0c2c36e1/latest/TRY").text
json_data = json.loads(request)
pprint(json_data["conversion_rates"])


dict_all_currencies=json_data["conversion_rates"]

dict_all_currencies_val=dict_all_currencies.values()

pprint(dict_all_currencies_val)


# to insert currency from api to ora19c below funct is used. 
def insert_currency_from_api(TRY,AED , ARS , AUD , BGN , BRL , BSD , CAD , CHF , CLP , CNY , COP , CZK , 
DKK , DOP , EGP , EUR , FJD , GBP , GTQ , HKD , HRK , HUF , IDR , ILS , INR , ISK , JPY , KRW , KZT ,
MVR , MXN , MYR , NOK , NZD , PAB , PEN , PHP , PKR , PLN , PYG , RON , RUB , SAR , SEK , SGD , THB , 
 TWD , UAH , USD , UYU , ZAR):
    """
    Insert a row to the billing_headers table
    :TRY: :AED:  :ARS:  :AUD:  :BGN:  :BRL:  :BSD:  :CAD:  :CHF:  :CLP:  :CNY:  :COP:  :CZK:
               :DKK:  :DOP:  :EGP:  :EUR:  :FJD:  :GBP:  :GTQ:  :HKD:  :HRK:  :HUF:  :IDR:  
               :ILS:  :INR:  :ISK:  :JPY:  :KRW:  :KZT:  
               :MVR:  :MXN:  :MYR:  :NOK:  :NZD:  :PAB:  :PEN:  :PHP:  :PKR:  :PLN:  :PYG: 
               :RON:  :RUB:  :SAR:  :SEK:  :SGD:  :THB:   :TWD:  :UAH:  :USD:  :UYU:  :ZAR:
    :return:
    """
    # construct an insert statement that add a new row to the billing_headers table
    sql = ('insert into halil_dba.currency_trx(TRY ,AED , ARS , AUD , BGN , BRL , BSD , CAD , CHF , CLP , CNY , COP , CZK , DKK , DOP , EGP , EUR , FJD , GBP , GTQ , HKD , HRK , HUF , IDR , ILS , INR , ISK , JPY , KRW , KZT ,MVR , MXN , MYR , NOK , NZD , PAB , PEN , PHP , PKR , PLN , PYG , RON , RUB , SAR , SEK , SGD , THB ,  TWD , UAH , USD , UYU , ZAR) '
        'values(:TRY ,:AED , :ARS , :AUD , :BGN , :BRL , :BSD , :CAD , :CHF , :CLP , :CNY , :COP , :CZK , :DKK , :DOP , :EGP , :EUR , :FJD , :GBP , :GTQ , :HKD , :HRK , :HUF , :IDR , :ILS , :INR , :ISK , :JPY , :KRW , :KZT , :MVR , :MXN , :MYR , :NOK , :NZD , :PAB , :PEN , :PHP , :PKR , :PLN , :PYG , :RON , :RUB , :SAR , :SEK , :SGD , :THB ,  :TWD , :UAH , :USD , :UYU , :ZAR)')

    try:
        # establish a new connection
        with cx_Oracle.connect("halil_dba", "halil_dba", "localhost:1522/orcl") as connection:
            # create a cursor
            with connection.cursor() as cursor:
                # execute the insert statement
                cursor.execute(sql, [TRY ,AED , ARS , AUD , BGN , BRL , BSD , CAD , CHF , CLP , CNY , COP , CZK , DKK , DOP , EGP , EUR , FJD , GBP , GTQ , HKD , HRK , HUF , IDR , ILS , INR , ISK , JPY , KRW , KZT ,MVR , MXN , MYR , NOK , NZD , PAB , PEN , PHP , PKR , PLN , PYG , RON , RUB , SAR , SEK , SGD , THB ,  TWD , UAH , USD , UYU , ZAR])
                # commit work
                connection.commit()
    except cx_Oracle.Error as error:
        print('Error occurred:')
        print(error)

        
insert_currency_from_api(*dict_all_currencies_val)
```
